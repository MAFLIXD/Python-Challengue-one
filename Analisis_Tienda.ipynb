### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

#1. Análisis de facturación



```python
ingresoTotal_tienda1 = tienda1["Precio"].sum()
print(ingresoTotal_tienda1)
```

```python
ingresoTotal_tienda2 = tienda2["Precio"].sum()
print(ingresoTotal_tienda2)
```

```python
ingresoTotal_tienda3 = tienda3["Precio"].sum()
print(ingresoTotal_tienda3)
```

```python
ingresoTotal_tienda4 = tienda4["Precio"].sum()
print(ingresoTotal_tienda4)
```

# 2. Ventas por categoría

# 4. Productos más y menos vendidos

```python
import matplotlib.pyplot as plt
import seaborn as sns

#GRAFICO 1 INGRESOS TOTALES

ventas = [ingresoTotal_tienda1,ingresoTotal_tienda2,ingresoTotal_tienda3,ingresoTotal_tienda4]
nombre = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

plt.figure(figsize=(5, 5)) # Opcional: ajusta el tamaño del gráfico
sns.barplot(x= nombre, y=ventas)
plt.title('Ingresos Totales por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Ingresos Totales')
plt.show()
```

```python
#GRAFICO 2 COSTO DE NEVIO DE CADA TIENDA

promedio_Envios = [promedioEnviosTienda1,promedioEnviosTienda2,promedioEnviosTienda3,promedioEnviosTienda4]
nombre = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

ventas_por_categoria_tienda1 = tienda1['Categoría del Producto'].value_counts()

plt.figure(figsize=(5, 5)) # Ajusta el tamaño del gráfico para un pastel
plt.pie(promedio_Envios, labels=nombre, autopct='%1.1f%%', startangle=140)
plt.title('Promedio costo de envio de cada tienda')
plt.axis('equal')  # Asegura que el gráfico de pastel sea un círculo.
plt.show()
```

```python
 # Asegúrate de que ya has calculado 'ventas_por_categoria_tienda1', 'ventas_por_categoria_tienda2', etc.
# Si no es así, ejecuta las celdas correspondientes en tu cuaderno.

# Combinar los datos de ventas por categoría en un solo DataFrame para facilitar la graficación
ventas_combinadas = pd.DataFrame({
    'Tienda 1': ventas_por_categoria_tienda1,
    'Tienda 2': ventas_por_categoria_tienda2,
    'Tienda 3': ventas_por_categoria_tienda3,
    'Tienda 4': ventas_por_categoria_tienda4
}).fillna(0) # Rellenar valores faltantes con 0 en caso de que alguna categoría no exista en una tienda

plt.figure(figsize=(10, 6)) # Ajusta el tamaño del gráfico si es necesario

for tienda in ventas_combinadas.columns:
    plt.plot(ventas_combinadas.index, ventas_combinadas[tienda], marker='o', label=tienda)

plt.title('Ventas por Categoría a través de las Tiendas')
plt.xlabel('Categoría del Producto')
plt.ylabel('Número de Ventas')
plt.xticks(rotation=45, ha='right') # Rota las etiquetas del eje X para mayor legibilidad
plt.legend()
plt.grid(True)
plt.tight_layout() # Ajusta el diseño para evitar que las etiquetas se superpongan
plt.show()
```
